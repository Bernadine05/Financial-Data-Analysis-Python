```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes = "true")
```

#IMPORTING OUR DATA AND SETTING PANDAS TO READ IT

```python
fin = pd.read_csv ("G:\My Drive\datasets\Finance_data.csv") 
fin.head(5)
```

```python
fin=fin.drop(["Mutual_Funds","Equity_Market", "Debentures", "Government_Bonds", "Fixed_Deposits", "PPF", "Gold"], axis =1)
fin.tail(5)
```

```python
#FINDING NULL VALUES
print(fin.isnull().sum())
```

```python
#FINDING DUPLICATES
duplicate_rows_fin=[fin.duplicated()]
print("number of duplicate rows:", duplicate_rows_fin)
```

#FINDING THE TOTAL NUMBER OF PEOPLE BY GENDER WHO ARE INVESTING

```python
sns.countplot(x="gender", data= fin, linewidth=3, edgecolor="blue")
plt.title("Number of people by gender")
plt.ylabel("Count")
plt.xlabel("Gender");
```

#IDENTIFYING THE NUMBER OF PEOPLE INVESTING BY AGE

```python
sns.countplot(x="age", data= fin, linewidth=3, edgecolor="blue")
plt.title("Number of people by Age Investing")
plt.ylabel("Count")
plt.xlabel("Age");
```

#FINDING RELATIONSHIP BETWEEN DIFFERENT FACTORS

```python
plt.figure(figsize=(20,10))

plt.subplot(1,2,1)
sns.countplot(x=fin["gender"], hue=fin["Investment_Avenues"], linewidth= 3, palette="Reds", edgecolor="blue")
plt.title("Number of Investment Avenues by Gender")

plt.subplot(1,2,2)
sns.countplot(x=fin["gender"], hue=fin["Stock_Marktet"], linewidth= 3, palette="Greens", edgecolor="blue") 
plt.title("Number of people by Gender Investing in Stock Market")              
```

#IDENTIFYING FACTORS THAT ARE MAINLY CONSIDERED WHILE MAKING INVESTMENT DECISONS

```python
sns.countplot(x="Factor", data= fin, linewidth=3, palette="YlOrRd", edgecolor="blue")
plt.title("Factors and how often they are Considered during Investment Decisions")
plt.ylabel("Count")
plt.xlabel("Factor");
```

#FINDING THE MEDIAN AGE OF PEOPLE INVESTING FOR VARIOUS PURPOSES IE WEALTH, SAVINGS FOR THE FUTURE AND RETURNS

```python
plt.figure(figsize=(16,8))
sns.pointplot(x="Purpose", y="age", data=fin, linestyles ="--", capsize=3, color="blue")
plt.title("Purpose for Investing and the median Age for each purpose");
```

```python
plt.figure(figsize=(20,10))

plt.subplot(1,2,1)
sns.countplot(x=fin["Duration"], hue=fin["gender"], linewidth=3, palette="bright", edgecolor="blue")
plt.title("DURATION OF INVESTMENT BY GENDER")

plt.subplot(1,2,2)
sns.countplot(x=fin["Invest_Monitor"], hue=fin["gender"], linewidth=3, palette="RdPu", edgecolor="blue")
plt.title("HOW OFTEN EACH GENDER INVESTS")
```
